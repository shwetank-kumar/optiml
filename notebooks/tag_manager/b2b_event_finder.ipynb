```python

def get_button_events(soup):

    # Find all button elements
    buttons = soup.find_all('button')

    # Find all input elements that are buttons
    input_buttons = soup.find_all('input', {'type': ['button', 'submit', 'reset']})

    # Combine the lists
    all_buttons = buttons + input_buttons

    # Remove all SVG elements from the buttons
    for btn in all_buttons:
        for svg in btn.find_all('svg'):
            svg.extract()

    button_events = []
    for idx, btn in enumerate(all_buttons):    
        # prompt2 = f"""Does the following html contain any of the following ecommerece web events - {ecommerce_events_string}
        # If it does then tell me which one and how did you come to the conclusion otherwise say it doesnt contain it:\n{btn}"""

        # prompt2 = f"""Does the following html contain any of the following ecommerece web events - {ecommerce_events_string}.
        # Your output is a dictionary with first key name {columns[0]} showing the html string, 
        # second key with key name {columns[1]} a boolean value showing if the event was found,
        # third key with key name {columns[2]} is the type of ecommerce event that was found if value of second key is true, 
        # fourth key with key name {columns[3]} is a short explanation of how you came to the conclusion.:\n{btn}"""

        prompt2 = f"""Does the following html contain any of the following ecommerece web events - {ecommerce_events_string}.
        Specifically, if anything in the html uses words similar to one of these events classify it as being associated with that event.
        Your output is a dictionary with first key name {columns[0]} showing the html string, 
        second key with key name {columns[1]} a boolean value showing if the event was found,
        third key with key name {columns[2]} is the type of ecommerce event that was found if value of second key is true, 
        fourth key with key name {columns[3]} is a short explanation of how you came to the conclusion.:\n{btn}"""

        # # Send the HTML content to the API using the chat model and ask to identify tags
        response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
                {"role": "system", "content": "You are a frontend engineer assistant at an ecommerce company."},
                {"role": "user", "content": prompt2 }
            ],
        temperature=0.1, 
        )
        event_tags = response.choices[0].message['content'].strip()
        print(idx, ": ", event_tags)
        print("------------------------------------------------------------")
        event_tags = json.loads(event_tags)
        button_events.append(event_tags)
        
    print(button_events)    
    return button_events

button_events = get_button_events(soup)

# Print events and tags 
# Open a file for writing
with open(events_csv, 'w', newline='') as events_file:
    writer = csv.DictWriter(events_file, fieldnames=columns)
    writer.writeheader()
    for button_event in button_events:
        writer.writerow(button_event)

```
